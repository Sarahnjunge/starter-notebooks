```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Dataset shape:", df.shape)

print("\nTrend direction distribution:")
print(df["trend_direction"].value_counts())

print("\nPercentage:")
print(df["trend_direction"].value_counts(normalize=True).round(3))
```



This is a **classification** problem.

The decision is to identify which content pages are likely to require attention so that editors can prioritize their review.

The model would classify each content page into categories such as declining or not declining based on historical performance signals. Classification is appropriate because the desired output is a discrete class rather than a continuous value or grouping.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
print(df[["trend_direction", "trend_pct"]].head())

print("\nUnique trend categories:")
print(df["trend_direction"].unique())

print("\nSummary of trend percentage:")
print(df["trend_pct"].describe())
```



The starter dataset contains **trend_direction**, which is derived from **trend_pct**.

Because the label is generated from a rule rather than being an observed future outcome, it should be treated as a proxy target. In a production system, I would prefer to predict an observed outcome measured in a future time window.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
print("Total content pages:", len(df))

print("\nTrend direction counts:")
print(df["trend_direction"].value_counts())

print("\nTrend direction percentages:")
print((df["trend_direction"].value_counts(normalize=True)*100).round(2))
```



I would use **Precision@50** as the primary success metric.

The business can only review a limited number of pages each week, so the model should ensure that the highest-ranked recommendations are genuinely valuable. A Precision@50 above 0.70 would indicate that most recommended pages truly deserve review.

The distribution of trend categories provides context for evaluating future model performance. Since only a small number of pages can be reviewed each week, prioritizing high-quality recommendations is more important than maximizing overall accuracy.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
print("Number of columns:", len(df.columns))

print("\nNumeric columns:")
print(df.select_dtypes(include="number").columns.tolist())

print("\nSummary statistics:")
display(df.describe())
```



A simple if-statement cannot effectively combine the many signals that influence content performance.

Search traffic depends on interactions between impressions, clicks, engagement, content characteristics, search position and freshness. Machine learning can learn these complex relationships automatically, making it more suitable than a single hand-written rule.
